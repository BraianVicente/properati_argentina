```python
def subtes(x):
    rango = 0.003
    latitud = x.lat
    longitud = x.lon
    coinciden = subte.loc[ (subte['Y'].between(latitud - rango, latitud + rango,inclusive=True)) & 
             (subte['X'].between(longitud - rango, longitud + rango, inclusive=True) ),:]
    if ( not(coinciden.empty) ):
        return coinciden['LINEA'].iloc[0]
    else:
        return 'N'
```
